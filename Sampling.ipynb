```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
# Load the voice signal
sampling_rate, voice_signal = wavfile.read('my_audio.wav')
```

```python
def sample_signal(signal, sample_rate):
    sample_interval = int(len(signal) / sample_rate)
    return signal[::sample_interval] 
```

```python
print("Choose a sampling rate:")
print("1. Nyquist Rate")
print("2. Half Nyquist Rate")
print("3. Above Nyquist Rate")
print("4. Custom Rate")
```

```python
# Get user input for sampling rate choice
choice = int(input("Enter your choice (1-4): "))

# Determine the sampling rates based on user choice
if choice == 1:
    nyquist_rate = 2 * (sampling_rate / 2)  # Original sampling rate divided by 2 gives max frequency
    sampled_signal = sample_signal(voice_signal, int(nyquist_rate))
elif choice == 2:
    half_nyquist_rate = sampling_rate / 2
    sampled_signal = sample_signal(voice_signal, int(half_nyquist_rate))
elif choice == 3:
    above_nyquist_rate = int(1.5 * (sampling_rate / 2))
    sampled_signal = sample_signal(voice_signal, above_nyquist_rate)
elif choice == 4:
    custom_rate = int(input("Enter your custom sampling rate: "))
    sampled_signal = sample_signal(voice_signal, custom_rate)
else:
    print("Invalid choice. Please select a number between 1 and 4.")
    sampled_signal = None
```

```python
# Plot the sampled signal if valid
if sampled_signal is not None:
    plt.figure(figsize=(12, 6))
    plt.plot(sampled_signal)
    plt.title(f'Sampled Signal at Rate Choice {choice}')
    plt.xlabel('Samples')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show() 
```
